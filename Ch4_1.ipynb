```python
import tkinter as tk
from tkinter import ttk

# GUI 창 생성하고 기본정보 설정
win = tk.Tk()
win.geometry("600x200")
win.title('업무 자동화 트릭')

# Test용 라벨 만들고 창에 올리기
lbTest = tk.Label(win, text = '라벨 테스트', width = 50)
lbTest.pack()

# Test용 콤보 박스 만들고 창에 올리기
cbTest = ttk.Combobox(win)

# 콤보 박스에서 선택 가능한 리스트 입력하기
cbTest['values'] = ['Text', '테스트', '시험']

# 콤보 박스에서 선택만 가능하게 강제하기
cbTest['state'] = 'readonly'
cbTest.pack()

# 연결 이벤트 발생시 동작 함수
def cbTest_changed(event):
    lbTest.config(text = cbTest.get())

# 콤보 박스의 값이 선택되었을 때 함수 호출
cbTest.bind('<<ComboboxSelected>>', cbTest_changed)

# GUI 반복
win.mainloop()
```

#### Check Button과 Radio Button 만들기

```python
import tkinter as tk
from tkinter import ttk

# GUI 창 생성하고 기본정보 설정
win = tk.Tk()
win.geometry("600x200")
win.title('업무 자동화 트릭')

# Test용 Check Button 만들고 창에 올리기
CheckVar1 = tk.IntVar()
CheckVar2 = tk.IntVar()
cbtnTest1 = tk.Checkbutton(win, text = 'Check Button 1', variable= CheckVar1 , width = 50)
cbtnTest2 = tk.Checkbutton(win, text = 'Check Button 2', variable= CheckVar2 , width = 50)
cbtnTest1.pack()
cbtnTest2.pack()

# Test용 Radio Button 만들고 창에 올리기
RadVar = tk.IntVar()
rbtnTest1 = tk.Radiobutton(win, text = 'Option 1', variable= RadVar , value = 1, width = 50)
rbtnTest2 = tk.Radiobutton(win, text = 'Option 2', variable= RadVar , value = 2, width = 50)
rbtnTest1.pack()
rbtnTest2.pack()

# GUI 반복
win.mainloop() 
```

```python
import tkinter as tk
from tkinter import ttk

# GUI 창 생성하고 기본정보 설정
win = tk.Tk()
win.geometry("600x200")
win.title('업무 자동화 트릭')

# Test용 Check Button 만들고 창에 올리기
CheckVar1 = tk.IntVar()
CheckVar2 = tk.IntVar()
cbtnTest1 = tk.Checkbutton(win, text = 'Check Button 1', variable= CheckVar1 , width = 50)
cbtnTest2 = tk.Checkbutton(win, text = 'Check Button 2', variable= CheckVar2 , width = 50)
cbtnTest1.pack()
cbtnTest2.pack()

# Test용 Radio Button 만들고 창에 올리기
RadVar = tk.IntVar()
rbtnTest1 = tk.Radiobutton(win, text = 'Option 1', variable= RadVar , value = 1, width = 50)
rbtnTest2 = tk.Radiobutton(win, text = 'Option 2', variable= RadVar , value = 2, width = 50)
rbtnTest1.pack()
rbtnTest2.pack()

# Test용 라벨 만들고 창에 올리기
lbCTest1 = tk.Label(win, text = 'Check Button1 테스트', width = 50)
lbCTest2 = tk.Label(win, text = 'Check Button2 테스트', width = 50)
lbRTest = tk.Label(win, text = 'Radio Button 테스트', width = 50)
lbCTest1.pack()
lbCTest2.pack()
lbRTest.pack()

# 연결 이벤트 발생시 동작 함수
def btnTest_click():
    lbCTest1.config(text = 'CheckVal1 값은 ' + str(CheckVar1.get()))
    lbCTest2.config(text = 'CheckVal2 값은 ' + str(CheckVar2.get()))
    lbRTest.config(text = 'RadVal 값은 ' + str(RadVar.get()))

# Test용 버튼 만들고 창에 올리기
btnTest = tk.Button(win, text = '값 가져오기 테스트', width = 50, command=btnTest_click )
btnTest.pack()

# GUI 반복
win.mainloop()
```
